# Fidelity heat maps

```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pandas as pd

df = pd.read_csv('../data/raw/heat_map_data.csv')
```

## Results

```python
width = df.loc[df['width'] == 25]
max_depth = width['depth'].max()
grouped = width.groupby(['depth', 'sdrp'])['fidelity'].mean().reset_index()
heatmap_data = grouped.pivot('depth', 'sdrp', 'fidelity')
```

```python
plt.rc('legend',fontsize=20)
plt.rcParams.update({'font.size': 20, 'lines.markersize': 12})
fig, ax = plt.subplots(figsize = (9, 9))
# plt.title("Fidelity vs. layer depth, 54 qubits, A100 GPU")
ax.set_facecolor("black")
plt.xlabel("SDRP")
plt.ylabel("Depth")

left = -0.00625
right = 1.00625
bottom = max_depth + 0.5
top = 0.5
extent = [left, right, bottom, top]

c = plt.colorbar(plt.imshow(heatmap_data, cmap='hot', interpolation='nearest', extent=extent, aspect=1.0125/(max_depth - 1), norm=colors.LogNorm()))
c.set_label('Fidelity', rotation=270, labelpad=20)
fig.axes[0].invert_yaxis()

fig.savefig('heat_map_fidelity_25.png', dpi=100)
```
