# Fidelity heat maps

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../data/raw/heat_map_data.csv')
```

## Results

```python
width = df.loc[df['width'] == 25]
max_depth = width['depth'].max()
grouped = width.groupby(['depth', 'sdrp'])['fidelity'].mean().reset_index()
heatmap_data = grouped.pivot('depth', 'sdrp', 'fidelity')
```

```python
fig, ax = plt.subplots(figsize = (9, 9))
# plt.title("Fidelity vs. layer depth, 54 qubits, A100 GPU")
plt.xlabel("SDRP")
plt.ylabel("Depth")

left = 0
right = 1
bottom = max_depth
top = 1
extent = [left, right, bottom, top]

plt.colorbar(plt.imshow(heatmap_data, cmap='hot', interpolation='nearest', extent=extent, aspect=1/(max_depth - 1)))

fig.savefig('heat_map_fidelity_25.png', dpi=100)
```
